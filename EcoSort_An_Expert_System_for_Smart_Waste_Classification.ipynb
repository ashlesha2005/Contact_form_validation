```python
# ✅ Install ipywidgets (only needed in Colab)
# Uncomment the next line if running in Google Colab
# !pip install ipywidgets

import random
import ipywidgets as widgets
from IPython.display import display
from difflib import get_close_matches

# Branded Introduction
print("Welcome to Smart Waste Classifier!")
print("Developed by: Ashlesha Parte ")


# Show Eco Tip of the Day
eco_tips = [
    "Use cloth bags instead of plastic.",
    "Start composting kitchen waste.",
    "Donate old clothes and toys.",
    "Recycle e-waste at proper centers.",
    "Avoid single-use plastics like straws and cups.",
    "Turn off lights when not in use.",
    "Reuse glass jars and containers.",
    "Use rechargeable batteries.",
    "Print only when necessary.",
    "Support local recycling programs."
]
print("Eco Tip of the Day:", random.choice(eco_tips), "\n")

# Waste Rulebase (50+ items)
waste_rules = {
    "banana peel": ("Wet Waste", "Green", "Can be composted at home."),
    "vegetable scraps": ("Wet Waste", "Green", "Use in compost for garden."),
    "fruit peel": ("Wet Waste", "Green", "Avoid throwing in plastic."),
    "leftover rice": ("Wet Waste", "Green", "Can feed animals or compost."),
    "tea bag": ("Wet Waste", "Green", "Great for composting."),
    "egg shell": ("Wet Waste", "Green", "Rinse before composting."),
    "bread": ("Wet Waste", "Green", "Can be composted if not moldy."),
    "flower garland": ("Wet Waste", "Green", "Use in vermicomposting."),
    "newspaper": ("Dry Waste", "Blue", "Recycle or donate to vendors."),
    "cardboard box": ("Dry Waste", "Blue", "Flatten before discarding."),
    "plastic bottle": ("Dry Waste", "Blue", "Reuse or recycle responsibly."),
    "glass bottle": ("Dry Waste", "Blue", "Handle with care."),
    "metal can": ("Dry Waste", "Blue", "Rinse before recycling."),
    "old clothes": ("Reusable", "None", "Donate to shelters or reuse."),
    "torn jeans": ("Reusable", "None", "Repurpose into bags or mats."),
    "school uniform": ("Reusable", "None", "Donate to NGOs."),
    "woolen sweater": ("Reusable", "None", "Share in winter donation drives."),
    "battery": ("Hazardous Waste", "Red", "Dispose at e-waste center."),
    "sanitary pad": ("Hazardous Waste", "Red", "Wrap and place in red bin."),
    "diaper": ("Hazardous Waste", "Red", "Wrap properly before discarding."),
    "used tissue": ("Hazardous Waste", "Red", "Do not mix with dry waste."),
    "medicine strip": ("Hazardous Waste", "Red", "Return to pharmacy or discard safely."),
    "syringe": ("Hazardous Waste", "Red", "Use medical waste bin."),
    "mask": ("Hazardous Waste", "Red", "Dispose responsibly."),
    "broken glass": ("Hazardous Waste", "Red", "Wrap before disposal."),
    "mobile phone": ("E-Waste", "Red", "Recycle at e-waste center."),
    "charger": ("E-Waste", "Red", "Avoid landfills – recycle."),
    "keyboard": ("E-Waste", "Red", "Donate or recycle."),
    "light bulb": ("E-Waste", "Red", "Dispose unbroken."),
    "remote": ("E-Waste", "Red", "Remove batteries before disposal."),
    "styrofoam": ("Dry Waste", "Blue", "Hard to recycle, avoid use."),
    "plastic bag": ("Dry Waste", "Blue", "Avoid usage, switch to cloth."),
    "aluminum foil": ("Dry Waste", "Blue", "Clean before discarding."),
    "toothpaste tube": ("Dry Waste", "Blue", "Empty and rinse."),
    "broken toy": ("Dry Waste", "Blue", "Donate if fixable."),
    "shoe": ("Dry Waste", "Blue", "Can be recycled or reused."),
    "matchbox": ("Dry Waste", "Blue", "Put in blue bin."),
    "old book": ("Reusable", "None", "Donate to school or libraries."),
    "expired food": ("Wet Waste", "Green", "Don't mix with plastic."),
    "burnt toast": ("Wet Waste", "Green", "Good for compost."),
    "thermocol": ("Dry Waste", "Blue", "Non-biodegradable."),
    "milk packet": ("Dry Waste", "Blue", "Wash before disposal."),
    "coconut shell": ("Dry Waste", "Blue", "Use for handicrafts."),
    "cooking oil packet": ("Dry Waste", "Blue", "Clean before disposal."),
    "detergent pouch": ("Dry Waste", "Blue", "Rinse before disposal."),
    "crackers": ("Hazardous Waste", "Red", "Avoid bursting. Dispose carefully."),
    "idol flowers": ("Wet Waste", "Green", "Use for compost."),
    "construction debris": ("Debris", "Gray", "Contact municipal service."),
    "brick": ("Debris", "Gray", "Do not throw in regular bins."),
    "cement bag": ("Debris", "Gray", "Can be reused.")
}

# Widgets
text_input = widgets.Text(
    value='',
    placeholder='Type your waste item here...',
    description='Type Item:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

waste_dropdown = widgets.Dropdown(
    options=[''] + sorted(waste_rules.keys()),
    description='Select Item:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

button = widgets.Button(description="Classify", button_style='success')
output = widgets.Output()

display(text_input)
display(waste_dropdown)
display(button, output)

def on_button_click(b):
    with output:
        output.clear_output()
        typed = text_input.value.strip().lower()
        selected = waste_dropdown.value.strip().lower()

        item = typed if typed else selected

        if item in waste_rules:
            waste_type, bin_color, tip = waste_rules[item]
            print(f"\nWaste Type: {waste_type}")
            print(f"Suggested Bin Color: {bin_color}")
            print(f"Eco Tip: {tip}")
        else:
            suggestions = get_close_matches(item, waste_rules.keys(), n=1, cutoff=0.6)
            if suggestions:
                closest = suggestions[0]
                print(f"⚠️ Did you mean '{closest}'?")
                waste_type, bin_color, tip = waste_rules[closest]
                print(f"\nWaste Type: {waste_type}")
                print(f"Suggested Bin Color: {bin_color}")
                print(f"Eco Tip: {tip}")
            else:
                print("⚠️ Item not found in knowledge base.")
                print("🌍 General Tip: Try to reduce, reuse, and recycle.")

button.on_click(on_button_click)
```
